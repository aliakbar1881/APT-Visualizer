```python
groups = groups.map(lambda i: str(i).strip())
techniques = techniques.map(lambda i: str(i).strip())
```

```python
filter_groups = []
for g in df.gid.unique():
    tactics = df[df.gid == g]['tactic'].value_counts().index
    if 'Execution' in tactics:
        if 'Persistence' in tactics:
            if 'Credential Access' in tactics:
                if 'Discovery' in tactics:
                    if 'Lateral Movement' in tactics:
                        filter_groups.append(g)
df = df[df.gid.isin(filter_groups)]
```
